```python
import pandas as pd
from sqlalchemy import create_engine

# Replace these variables with your actual MySQL connection details
user = 'root'
password = 'admin'
host = '127.0.0.1'
port = '3306'
database = 'coffeeshop'

# Create the SQLAlchemy engine with the correct format
engine = create_engine(f'mysql+mysqlconnector://{user}:{password}@{host}:{port}/{database}')

# Read the tables into DataFrames
customer = pd.read_sql('SELECT * FROM customer', con=engine)
product = pd.read_sql('SELECT * FROM product', con=engine)
order = pd.read_sql('SELECT * FROM order_details', con=engine)
supplier = pd.read_sql('SELECT * FROM supplier', con=engine)
employee=pd.read_sql('SELECT * FROM employee', con=engine)


# Printing the first 5 records from each table
print("Customer Data:")
print(customer.head())

print("\nProduct Data:")
print(product.head())

print("\nOrder Data:")
print(order.head())

print("\nSupplier Data:")
print(supplier.head())

print("\nemployee Data:")
print(employee.head())
```

# **Total Revenue Generated from Orders Product-wise**

# **Month-wise Total Sales**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert order_date to datetime if it's not already in datetime format
order['order_date'] = pd.to_datetime(order['order_date'])

# Extract the month from the order date
order['month'] = order['order_date'].dt.month

# Group the orders by month and calculate total sales for each month
monthly_sales = order.groupby('month')['total_price'].sum().reset_index()

# Plotting month-wise total sales
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['month'], monthly_sales['total_price'], marker='o', color='skyblue')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Month-wise Total Sales')
plt.xticks(range(1, 13))
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **Month and Year-wise Total Sales**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert order_date to datetime if it's not already in datetime format
order['order_date'] = pd.to_datetime(order['order_date'])

# Extract the month and year from the order date
order['month'] = order['order_date'].dt.month
order['year'] = order['order_date'].dt.year

# Group the orders by month and year and calculate total sales for each month-year combination
monthly_sales = order.groupby(['year', 'month'])['total_price'].sum().reset_index()

# Plotting month and year-wise total sales
plt.figure(figsize=(10, 6))
for year in monthly_sales['year'].unique():
    sales_year = monthly_sales[monthly_sales['year'] == year]
    plt.plot(sales_year['month'], sales_year['total_price'], marker='o', label=year)

plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Month and Year-wise Total Sales')
plt.xticks(range(1, 13))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **Daily Order Count**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert order_date to datetime if it's not already in datetime format
order['order_date'] = pd.to_datetime(order['order_date'])

# Extract date from the order date
order['order_day'] = order['order_date'].dt.date

# Group the orders by order date and count the number of orders for each date
daily_order_count = order.groupby('order_day').size().reset_index(name='order_count')

# Find the peak order date
peak_order_date = daily_order_count[daily_order_count['order_count'] == daily_order_count['order_count'].max()]

# Plotting daily order count
plt.figure(figsize=(10, 6))
plt.plot(daily_order_count['order_day'], daily_order_count['order_count'], color='skyblue')
plt.scatter(peak_order_date['order_day'], peak_order_date['order_count'], color='red', label='Peak Order Date')
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.title('Daily Order Count')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print("Peak Order Date:")
print(peak_order_date)
```
